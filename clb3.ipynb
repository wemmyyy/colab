fazer um programa que o usuario ira entrar com 04 notas, vamos calcular a media e mostrar o usuario a media

```python
nota1 = int(input("digite a primeira nota: "))
nota2 = int(input("digite a segunda nota: "))
nota3 = int(input("digite a terceira nota: "))
nota4 = int(input("digite a quarta nota: "))

#processamento

calculo = nota1 + nota2 + nota3 + nota4
calculo2 = calculo / 4

#saida

print(f"a media do calculo das notas é igual a: {calculo2}")
```

fazer um cadastro simples de usuario

objetivo: usar str (string)

logica

entrada: nome e idade, Processamento: armazenar, Saida: exibir uma mensagem "ola (nome)", voce tem x anos

```python
#entrada

nome = str(input("qual seu nome?: "))
idade = int(input("quantos anos vc tem?: "))

#saida

print(f"ola {nome}, voce tem {idade} anos.")
```
