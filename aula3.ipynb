```python
#operadores aritimeticos

#pedir ao usuario para digitar 02 numeros

#entrada

a = int(input("digite um numero: "))
b = int(input("digite um outro numero: "))

#processamento

#soma

soma = a + b
soma2 = a - b
soma3 = a * b
soma4 = a / b
div_int = a // b
resto = a % b
potencia = a ** b
#saida

print("")
print("----------------OPERADORES ARITIMETICOS------------------")
print("")
print("-------------------------SOMA----------------------------")
print("")
print(f"o resultado da soma de {a} mais {b} é igual a: {soma}")
print("")
print("-----------------------subtração-------------------------")
print("")
print(f"o resultado da subtração de {a} menos {b} é igual a: {soma2}")
print("")
print("-----------------------multiplicação---------------------")
print("")
print(f"o resultado da multiplicação de {a} vezes {b} é igual a: {soma3}")
print("")
print("-----------------------divisão---------------------------")
print("")
print(f"o resultado da divisão de {a} por {b} é igual a: {soma4}")
print("")
print("--------------------divisão inteira-----------------------")
print("")
print(f"o resultado da divisão inteira de {a} por {b} é igual a: {div_int}")
print("")
print("--------------------RESTO-----------------------")
print("")
print(f"o resultado da divisãoO de {a} dividido {b} o resto é igual a: {resto}")
print("")
print("--------------------POTÊNCIA-----------------------")
print("")
print(f"o resultado da POTENCIAÇÃO de {a} potenciação de {b} é igual a: {potencia}")
print("")
print("----------------------------------------------------------")
```
